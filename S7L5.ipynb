```python
#grafici
#mi dava errore poichÃ¨ le colonne normalizzate dovevano essere aggiunte al dataframe. quando abbiamo fatto questo
#passaggio, Ã¨ possibile procedere con i grafici.

import matplotlib.pyplot as plt
df['z_score_temperatura_media'] = (df['temperatura_media'] - df['temperatura_media'].mean()) / df['temperatura_media'].std()
df['z_score_precipitazioni'] = (df['precipitazioni'] - df['precipitazioni'].mean()) / df['precipitazioni'].std()
df['z_score_umidita'] = (df['umidita'] - df['umidita'].mean()) / df['umidita'].std()
df['z_score_velocita_vento'] = (df['velocita_vento'] - df['velocita_vento'].mean()) / df['velocita_vento'].std()

plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.hist(df['z_score_temperatura_media'], bins=20, color='blue', alpha=0.7, ec= "black", fc= "orange")
plt.title('temperatura_media (z-score)')
plt.xlabel('z-score')
plt.ylabel('frequenza')


plt.subplot(2, 2, 2)
plt.hist(df['z_score_precipitazioni'], bins=20, color='green', alpha=0.7, ec= "black", fc= "green")
plt.title('precipitazioni (z-score)')
plt.xlabel('z-score')
plt.ylabel('frequenza')

plt.subplot(2, 2, 3)
plt.hist(df['z_score_umidita'], bins=20, color='red', alpha=0.7, ec= "black", fc= "red")
plt.title('umiditÃ  (z-score)')
plt.xlabel('z-score')
plt.ylabel('frequenza')

plt.subplot(2, 2, 4)
plt.hist(df['z_score_velocita_vento'], bins=20, color='purple', alpha=0.7, ec= "black", fc= "purple")
plt.title('velocita_del_vento(z-score)')
plt.xlabel('z-score')
plt.ylabel('frequenza')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calcoliamo la matrice di correlazione
corr_matrix = df[['temperatura_media', 'precipitazioni', 'umidita', 'velocita_vento']].corr()

# Creiamo la heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlazione tra variabili meteorologiche', fontsize=16)
plt.show()
```
